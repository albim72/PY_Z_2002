import bibliotek: **numpy** oraz **matplotlib**

```python
import numpy as np
import matplotlib.pyplot as plotter
```

```python
samplingFrequency = 100
samplingIntrval = 1/samplingFrequency
beginTime = 0
endTime = 10
signal1Freq = 4
signal2Freq = 7
```

```python
time = np.arange(beginTime,endTime,samplingIntrval)
amplituda1 = np.sin(2*np.pi*signal1Freq*time)
amplituda2 = np.sin(2*np.pi*signal2Freq*time)
```

```python
figure,axis = plotter.subplots(4,1)
plotter.subplots_adjust(hspace=1)
axis[0].set_title('fala o częstotliwości 4Hz')
axis[0].plot(time,amplituda1)
axis[0].set_xlabel('Czas')
axis[0].set_ylabel('Amplituda')

axis[1].set_title('fala o częstotliwości 7Hz')
axis[1].plot(time,amplituda2)
axis[1].set_xlabel('Czas')
axis[1].set_ylabel('Amplituda')

amplituda = amplituda1 + amplituda2
axis[2].set_title('nałożenie fal')
axis[2].plot(time,amplituda)
axis[2].set_xlabel('Czas')
axis[2].set_ylabel('Amplituda')

#reprezentacja funkcji falowej z użyciem transformaty Fouriera

fourierTransform = np.fft.fft(amplituda)/len(amplituda) #normalizacja amplitudy
fourierTransform = fourierTransform[range(int(len(amplituda)/2))] #wybór próbki
tpc = len(amplituda)
values = np.arange(int(tpc/2))
timePeriod = tpc/samplingFrequency
frequencies = values/timePeriod

axis[3].set_title('transformata Fouriera dla wybranych częstotliwości')
axis[3].plot(frequencies,abs(fourierTransform))
axis[3].set_xlabel('Częstotliwości')
axis[3].set_ylabel('Amplituda')

plotter.show()
```
